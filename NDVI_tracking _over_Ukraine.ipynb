```
from pyhdf.SD import SD, SDC  # Import the HDF library

# Reading the file:
path = "MOD13Q1.A2022177.h20v04.061.2022195121012.hdf"
file = SD(path, SDC.READ)

# Printing all the datasets names
datasets_dic = file.datasets()
for idx,sds in enumerate(datasets_dic.keys()):
    print (idx,sds)
    
```

```
FILE_NAME = 'MOD13Q1.A2022177.h20v04.061.2022195121012.hdf'
hdf = SD(FILE_NAME, SDC.READ)

# List available SDS datasets.
print (hdf.datasets())
```

```
DATAFIELD_NAME="250m 16 days NDVI"
data2D = hdf.select(DATAFIELD_NAME)
data = data2D[:,:]
```

```
from matplotlib import pyplot as plt
plt.imshow(data)
plt.colorbar()
plt.show()
```
